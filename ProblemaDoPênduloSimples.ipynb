### Gráficos de  x * t para os ítens 1, 2, 3 e 4. Calcule o valor do período no caso de pequenos ângulos e comente o que ocorre com o período a medida que o ângulo inicial aumenta.

#### Plotando x * v para os ítens 1, 2 ,3 e 4

### Gráficos da energia total para cada item e compare com o valor que se espera analiticamente.

```python
thetalist = [0.1,0.5,1.,3.,3*3.14159,-3*3.14159]
v0s = [0,0,0,0,-0.1,0.1]
colorplot = ["b","g","r","c","y","m","b","g","r","c","y","m"]

counter = 0

for i in thetalist:

  m = 1.
  g = 10.
  l = 10.
  theta_zero = i
  t = 0.
  delta = 0.1
  tf = 24
  v0 = v0s[counter] 
  vel = v0
  omega = g/l

  plot_ec = []
  plot_eg = []
  plot_em = []
  plot_tempo = [t]

  theta = theta_zero

  EtAnalitica = (m*g*l*(1-cos(theta))) + (0.5*m*(l**2)*(vel**2))
  plot_ema = []

  #Euler-Cromer
  while (t<tf):
      
      plot_ema.append(EtAnalitica)

      Ec = 0.5*m*(l**2)*(vel**2)
      plot_ec.append(Ec)
      
      Eg = m*g*l*(1-cos(theta))
      plot_eg.append(Eg)
      
      Em = Ec + Eg
      plot_em.append(Em)

      theta = theta + vel*delta

      vel = vel - omega * sin(theta) * delta      

      t = t + delta
      plot_tempo.append(t)
  
  #Plotando X

  del plot_tempo[-1] #deixando as listas do mesmo tamanho

  #Analíticamente
  '''
  De um ponto de vista anaítico, Em = Ec + Eg, podemos escolher o tempo t0 arbitáriamente
  pelo motivo de termos a menor velocidade do sistema, nos primeiros casos é 0 , 
  logo a Energia Mecânica total do sistema é a própria
  Energia Potencial Graviational para os 4 primeiros casos , e para os ultimos 2
  teremos um singela mudança, pois apesar de existir, a velocidade inicial é pequena
  ''' 
  #plt.plot(plot_tempo, plot_ec, label=f"Energia Cinética",c = colorplot[counter])

  #plt.plot(plot_tempo, plot_eg, label=f"Energia potencial Gravitacional",c = colorplot[counter+1])

  plt.plot(plot_tempo, plot_em, label=f"Energia Total Theta={i}",c = colorplot[counter+2])
  plt.plot(plot_tempo,plot_ema, label="Energia Total Analitica", c="black", linestyle="--")

  plt.ylabel("Energia")
  plt.xlabel("Tempo")
  plt.legend()

  plt.show()

  counter = counter + 1
```
