### !!Pode-se fazer tabelas com html (pesquisar)

```python
import matplotlib.pyplot as plt

def pot2(x):
    y = x**2
    return y
    
x = [-3, -2, -1, 0, 1, 2, 3]
h = [-10, -7, -4, 0, 6, 8, 9]

z = list(map(pot2,x+h))

print(z)

plt.plot(x,h)
plt.show()
```
