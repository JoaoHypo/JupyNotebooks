# Função média, desvio padrao, e desvio padrao da média

## Lista > [                                ] > N

```python
import random as rnd
from math import sqrt
```

```python
rnd.random()  #Por padrão imprime somente números intervalados de 0 a 1
```

```python
N = 100
lista = [ ]
for i in range(N):
    lista.append(rnd.random())
print(lista)
```

```python
S = 0.0
S2 = 0.0

N = len(lista)
for i in range(N):
    S = S + lista[i]
    S2 =  S2 + (lista[i]**2)
S = S/N
S2 = S2/N -  S**2
sigma =  sqrt(S2)
delta = sigma/sqrt(N)
```

```python
print('Valor médio:', S)
print('Valor Variância:', S2)
print('Valor erro na média:', delta)
```
